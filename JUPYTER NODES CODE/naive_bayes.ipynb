```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn import metrics
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.metrics import precision_recall_curve
from sklearn.metrics import confusion_matrix
from sklearn.metrics import f1_score
```

```python
df=pd.read_csv("User_Data.csv")
df
```

```python
df.drop(columns = ['User ID'],axis=1,inplace=True)  #inplace is used to fix the changes or permannent the changes
df
```

```python
#Label encodind
le =LabelEncoder()
df['Gender']=le.fit_transform(df['Gender']) #it converts the gender in (o and  1),it is onehot encoding
df
```

```python
#split data into dependent /independent variable
X=df.iloc[:, :-1].values #it represent the rows and colums the -1 is represented as coloums (when we dont know how many coloms are their)
Y=df.iloc[:, -1].values
```

```python
#split data into test/train set
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size =0.25,random_state =True)
```

```python
#scale dataset
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
Y_test = sc.transform(X_test)
```

```python
#classifier
classifier = GaussianNB()
classifier.fit(X_train,Y_train)
```

```python
#pridiction
y_pred = classifier.predict(X_test)
#accuracy
accuracy_score(Y_test,y_pred)
```
