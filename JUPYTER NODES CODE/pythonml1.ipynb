```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
#plot a function
def plot_function(func):
    x_values = np.linspace(-10,10,1000)     #linspace has 3 parameters (startingpoint,endingpoint,number of values in between)
    y_values = [func(x) for x in x_values]   # list comprehention is used to make our code to compact an consice, this function will be computed for every value of x in x_values
    # print("x values are ",x_values)
    # print("y values are ",y_values)
    plt.plot(x_values,y_values)
    
plot_function(lambda x:x**2)    #-> lambda funtions are the anonymus functions with out a name and the definition of that function is written in a single line.(you can pass it want you want in x**2),for plot the sinx graph to write np.sinx.
```

```python
def plot_derivative(func):
    x_values =np.linspace(-5,5,1000)   #1000 is no.of itrations
    delta_x=0.0001
    y_values=(func(x_values + delta_x) - func(x_values))/delta_x    #finding derivative
    plt.plot(x_values,y_values)
plot_derivative(lambda x:np.sin(x))
```

```python
#Q)Given a function f and i straight point w(weight),
#Q)try to find a minima  or gradient descent
#Q)when will  call the function it will keep in on ocillation
def gradient_descent(func,w):
    list_of_weights = []
    weight = w
    delta=0.0001
    for i in range(1000):
        derivative=(func(weight+delta)-func(weight))/delta
        weight = weight - derivative
        list_of_weights.append(weight)
    return list_of_weights
gradient_descent(lambda x: x**2+4*x+3, 10)
```

```python
def gradient_descent(func,w):
    list_of_weights = []
    weight = w
    delta=0.0001
    learning_rate=0.1
    for i in range(1000):
        derivative=(func(weight+delta)-func(weight))/delta
        weight = weight - (learning_rate*derivative)
        list_of_weights.append(weight)
    return list_of_weights
gradient_descent(lambda x: x**2+4*x+3, 10)
```
